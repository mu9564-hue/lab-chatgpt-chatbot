## 0) 자가진단 — 이 셀이 읽히나요?


```python
def summarize(reviews: list[dict]) -> dict:
    total = len(reviews)
    positives = [r for r in reviews if r["rating"] >= 3]   # 컴프리헨션 + 필터
    avg = sum(r["rating"] for r in reviews) / total if total else 0
    return {"count": total, "positive": len(positives), "avg": round(avg, 2)}

data = [{"rating": 5}, {"rating": 2}, {"rating": 4}]
print(summarize(data))   # {'count': 3, 'positive': 2, 'avg': 3.67}

```

## 1) 값 · 자료구조
리뷰 한 건 = `dict`, 모음 = `list`. LLM 의 `messages` 가 바로 이 형태.


```python
name, age, ratio = "리뷰", 10_000, 0.95      # 다중 할당
print(f"{name} {age}건 ({ratio:.0%})")        # f-string 포매팅

xs = [5, 2, 4, 1]
xs.append(3)
print(xs[0], xs[-1], xs[1:3])                  # 인덱싱 / 슬라이싱
print(sorted(xs, reverse=True))                # [5,4,3,2,1]
```

## 2) 흐름 · 함수 · 컴프리헨션
API 호출은 전부 **함수 + 키워드 인자** 모양. 전처리는 **컴프리헨션** 사고.


```python
xs = [5, 2, 4, 1, 3]
for i, x in enumerate(xs):          # 인덱스 + 값
    if x >= 4:
        print(i, "긍정")
    elif x == 3:
        continue
    else:
        pass
```

## 4) pandas 30초 맛보기 — 실습 1 의 사전 준비

실습 1 부터 **pandas** 를 쓴다. 핵심은 둘뿐:
- `DataFrame` = 표(엑셀 시트), `Series` = 그 표의 **한 컬럼**(=list 의 표 버전)
- 컬럼 전체에 문자열 연산은 `.str` 을 붙여 한 번에 — `s.str.replace(...)` (반복문 불필요)

```python
import pandas as pd

df = pd.DataFrame({"text": [" 좋아요 ", "[광고] 별로", "URL http://x.co 포함"],
                   "rating": [5, 2, 4]})

# Series 1개에 문자열 연산을 통째로 (반복문 없이)
clean = df["text"].str.strip().str.replace(r"https?://\S+", "", regex=True)
print(clean.tolist())

# 불리언 필터 + groupby (실습 1·2 에서 그대로 쓴다)
print("광고 포함 행 수:", df["text"].str.contains(r"\[광고\]").sum())
print(df.groupby("rating")["text"].count())
```
